```python
#importing matplotlib
import matplotlib.pyplot as plt
```

import NUMPY for getting data for plot


```python
import numpy as np
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

Plotting the data


```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(y,z)
plt.show()
```

```python
plt.plot(x,z)
plt.show()
```

```python
#adding axis name for x and y-axis
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('sine wave')
plt.show()
```

```python
#parabola curve
x=np.linspace(-10,10,20)
y=x**2
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('parabola')
plt.plot(x,y)
plt.show()
```

```python
#to plot with unique symbol
#plt.plot(x,y,'+') # simple + symbol only
plt.plot(x,y,'r+') # + symbol  with read color
plt.show()
```

```python
plt.plot(x,y,'g*')
plt.show()
```

```python
#two equation line in one graph
x=np.linspace(-5,5,50)
plt.plot(x,np.sin(x),'g-')
plt.plot(x,np.cos(x),'r--')
plt.show()
```

BAR PLOT

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
langs=['c','c++','java','python','php']
students=[23,17,35,29,12]
ax.bar(langs,students)
plt.xlabel('languages')
plt.ylabel('number of students')
plt.title('bar graph')
plt.show()
```

Pie Chard

```python
fig1=plt.figure()
ax1=fig1.add_axes([0,0,1,1])
langs=['c','c++','java','python','php']
students=[23,17,35,29,12]
ax1.pie(students,labels=langs)
plt.show()
```

Scatter Plot

```python
x=np.linspace(0,10,20)
y=np.sin(x)
z=np.cos(x)
fig2=plt.figure()
ax2=fig2.add_axes([0,0,1,1])
ax2.scatter(x,y)
plt.show()
```

3d scatter plot

```python
fig3=plt.figure()
ax=plt.axes(projection='3d')
z=np.linspace(0,15,1000)
x=np.sin(z)
y=np.cos(z)
ax.scatter3D(x,y,z,c=z,cmap='Greens')
plt.show()
```
